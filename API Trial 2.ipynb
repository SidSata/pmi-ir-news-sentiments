```python
# Removing punctuation
import re
def preprocessor(text):
    if not text:
        return ""
    text = re.sub('<[^>]*>', '', text)
    emoticons = re.findall('(?::|;|=)(?:-)?(?:\)|\(|D|P)', text)
    text = re.sub('[\W]+', ' ', text.lower()) +\
        ' '.join(emoticons).replace('-', '')
    return text

original_df = article_df.copy(deep=True)

article_df['abstract'] = article_df['abstract'].apply(preprocessor)
article_df['lead_paragraph'] = article_df['lead_paragraph'].apply(preprocessor)
article_df['headline'] = article_df['headline'].apply(preprocessor)
```

```python
# Tokenizing documents using porter stemmer
from nltk.stem.porter import PorterStemmer

porter = PorterStemmer()

def tokenizer(text):
    return text.lowercase().split()


def tokenizer_porter(text):
    lst = [porter.stem(word) for word in text.split()]
    new_str = ""
    for i in lst:
        new_str+=i+' '
    return new_str[0:-1]
# 
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from nltk.corpus import stopwords

stop = stopwords.words('english')
stop = list(map(tokenizer_porter, stop))

def tf_idf_mapper(docs):
    count = CountVectorizer()
    bag = count.fit_transform(docs)
    
    tfidf = TfidfTransformer(use_idf=True, norm=None, smooth_idf=True)
    raw_tfidf = tfidf.fit_transform(count.fit_transform(docs)).toarray()[-1]
    
    l2_tfidf = (raw_tfidf+0.01) / (np.sqrt(np.sum(raw_tfidf**2))+0.01)
    
    mappings = {}
    for key in count.vocabulary_:
        val = count.vocabulary_[key]
        term = l2_tfidf[val]
        mappings[key] = term
    return mappings



def get_sentiment_score(docs):
    mappings = tf_idf_mapper(docs)
    score = 0
#     scores = []
    for word in mappings:
        if word in words_and_scores:
#             print(word, words_and_scores[word])
            score += mappings[word]* words_and_scores[word]
#             scores.append(mappings[word]*words_and_scores[word])
#     print(scores)
#     print(sum(scores))
    return score
    
def get_sentiment_articles(index):
    docs = np.array([
        article_df['abstract'].iloc[index],
        article_df['lead_paragraph'].iloc[index],
        article_df['headline'].iloc[index]])
    new_docs = []
    for sen in docs:
        words = sen.split()
        ns = ''
        for w in words:
            if w not in stop:
                ns += w+' '
        new_docs.append(ns[:-1])
    docs = np.array(new_docs)
#     print(docs)
#     fields = ['abstract', 'lead_paragraph', 'headline']
#     docs = []
#     for curr in fields:
#         if article_df[curr].iloc[i] != None:
#             docs.append(article_df[curr].apply(tokenizer_porter).iloc[i])
#     docs = np.array(docs)
    return get_sentiment_score(docs)

article_df['abstract'] = article_df['abstract'].apply(tokenizer_porter)
article_df['lead_paragraph'] = article_df['lead_paragraph'].apply(tokenizer_porter)
article_df['headline'] = article_df['headline'].apply(tokenizer_porter)
```
